```python
import requests
import json
import math
import time  

URL = "http://localhost:11434/api/generate"
MODEL = "llama3.2:1b"

payload = {
    "model": MODEL,
    "prompt": "How big is the universe?",
    "stream": False, 
    "logprobs": True,
    "top_logprobs": 3
}

print(f"Sending prompt to {MODEL}...\n")

try:
    start_time = time.time() # add a timer to see how long Ollama takes
    
    response = requests.post(URL, json=payload)
    response.raise_for_status() # Check for basic HTTP errors
    
    end_time = time.time()
    
    # Calculate the total time taken
    duration = end_time - start_time
    
    data = response.json()
    
    print("-" * 50)
    print(f"PYTHON TIMER: Request completed in {duration:.2f} seconds")
    print("-" * 50)
    
    print("\nFULL RESPONSE:")
    print("-" * 50)
    print(data.get("response", "No response text found."))
    
    print("\n" + "-" * 50)
    print("TOKEN PROBABILITIES (First 5 tokens):")
    print("-" * 50)
    
    # extract logprobs
    logprobs_data = data.get("logprobs", [])
    
    # print some tokens
    for i, token_data in enumerate(logprobs_data[:5]):
        print(f"\nGeneration Step {i + 1}:")
        
        for candidate in token_data.get("top_logprobs", []):
            token = candidate.get("token", "")

            token_clean = token.replace("\n", "\\n") 
            logprob = candidate.get("logprob", 0.0)
            
            percentage = math.exp(logprob) * 100
            
            print(f"  -> '{token_clean}': Logprob {logprob:.4f}  ({percentage:.2f}%)")

    
    if "total_duration" in data:
        internal_duration = data["total_duration"] / 1_000_000_000
        print("\n" + "-" * 50)
        print(f"OLLAMA INTERNAL TIMER: Processing took {internal_duration:.2f} seconds")
        print("-" * 50)

except requests.exceptions.ConnectionError:
    print("Error: Could not connect to Ollama. Is the Ollama app running in the background?")
except Exception as e:
    print(f"An error occurred: {e}")
```
