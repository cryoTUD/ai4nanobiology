## 2-D Fourier Analysis

Now that we considered 1D signals (= signals with one independent variable) and how we can use Fourier theory to analyse those signals, we can also consider signals that are more complicated. For example we can consider 2D images e.g. those taken by a digital camera, which we can represent by brightness varying as a function of two spatial vaiable: the coordinates (x,y). 

### Image as a 2D Signal
What do we mean by saying that a digital image can be considered a 2D signal? You know that a digital image is composed of small discrete units, or pixels. Thus in essence we mean that we can represent an image by a 2D matrix of numbers, each representing the brightness of a pixel located at position x,y in our image. We can actually look at this matrix in any data analysis software tool like Python, Excel or Matlab. 
    
Run the following interactive plot to see the numbers yourself. By changing the shift x and shift y values you can move a small rectangle around the displayed image and and investigate how the brightness varies. Try to choose a location that is located at the edge between high and low brightness to see this best.
