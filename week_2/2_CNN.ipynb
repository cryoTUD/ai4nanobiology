## 1. What is a CNN?

<!-- Pointer:
     - CNN is an FC network with two structural constraints:
       1. Sparse connectivity (locality): each output unit looks at a small spatial window
       2. Weight sharing: the same kernel is applied at every spatial location
     - These constraints encode two priors about images:
       1. Nearby pixels are related; far pixels are mostly not
       2. A feature (edge, corner, texture) means the same thing wherever it appears
     - Connect back: this is an inductive bias matched to image data
-->
