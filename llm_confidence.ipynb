```python
import transformers
from transformers import BloomForCausalLM
from transformers import BloomTokenizerFast
import torch
```

```python
model = BloomForCausalLM.from_pretrained("bigscience/bloom-1b7")
tokenizer = BloomTokenizerFast.from_pretrained("bigscience/bloom-1b7")
```

```python
prompt = "What is the capital of France? a) Paris b) London c) Berlin d) Madrid. The answer is: "

result_length = 25
inputs = tokenizer(prompt, return_tensors="pt")
```

```python
# Greedy Search
print(tokenizer.decode(model.generate(inputs["input_ids"], 
                       max_length=result_length
                      )[0]))
```

```python
# Beam Search
print(tokenizer.decode(model.generate(inputs["input_ids"],
                       max_length=result_length, 
                       num_beams=2, 
                       no_repeat_ngram_size=2,
                       early_stopping=False
                      )[0]))
```

```python
# Sampling Top-k + Top-p
print(tokenizer.decode(model.generate(inputs["input_ids"],
                       max_length=result_length, 
                       do_sample=True, 
                       top_k=50, 
                       top_p=0.9
                      )[0]))
```

```python
prompt = "Global warming is the current rise in temperature of the air and oceans. It is happening mainly because humans burn coal, oil, and natural gas; and cut down forests. Average temperatures today are about 1 °C (1.8 °F) higher than before people started burning a lot of coal around 1750. In some parts of the world it is less and some more. Most climate scientists say that by the year 2100 temperatures will be 2 °C (3.6 °F) to 4 °C (7.2 °F) higher than they were before 1750. \nGlobal warming is a significant and rapidly accelerating issue caused by human activities, such as greenhouse gas emissions, deforestation, and the burning of fossil fuels. It is having a profound impact on the environment and our planet. The present global warming is mostly because of people burning things, like gasoline for cars and natural gas to keep houses warm. But the heat from the burning itself only makes the world a tiny bit warmer: it is the carbon dioxide from the burning which is the biggest part of the problem. Among greenhouse gases, the increase of carbon dioxide in the atmosphere is the main cause of global warming. Svante Arrhenius predicted this more than a hundred years ago. Arrhenius confirmed the work of Joseph Fourier 200 years ago."

print(len(prompt.split()))
```

```python
input_tokens_for_summarisation = tokenizer.encode("Summarise this text: " + prompt, return_tensors="pt")
model_output_ids = model.generate(input_tokens_for_summarisation)
print(tokenizer.decode(model_output_ids[0], skip_special_tokens=True))
```
